```python
import numpy as np
import pandas as pd 
import json
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv') 
```

```python
movies= movies.merge(credits,on='title')
```

```python
movies["genres"] = movies["genres"].apply(lambda x: [g["name"] for g in json.loads(x)])
```

```python
movies
```

```python
movies["keywords"] = movies["keywords"].apply(lambda x: [g["name"] for g in json.loads(x)])
```

```python
movies
```

```python
movies["cast"] = movies["cast"].apply(lambda x: [g["name"] for g in json.loads(x)][:3])
```

```python
movies
```

```python
def fetch_director(text):
    L = []
    for i in json.loads(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.loc[:, 'cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies.loc[:, 'crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies.loc[:, 'keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies.loc[:, 'genres'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
movies.head()
```

```python
movies.loc[:,'overview'] = movies['overview'].apply(lambda x: str(x).split() if pd.notnull(x) else [])

```

```python
movies.head(20)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.tags[0]
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
new['tags'][0]
```

```python
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x:x.lower())
```

```python
new.head()
```

```python
new['tags'][0]
```

```python
import nltk
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new['tags'] = new['tags'].apply(stem)
```

```python
new['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
    
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity.shape
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Man of Steel')
```
